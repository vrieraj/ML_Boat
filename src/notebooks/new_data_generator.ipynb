```python
import pandas as pd
import numpy as np
from faker import Faker
import random  
```

```python
fake = Faker('en_UK')
n = 300


df = pd.DataFrame({
    "PassengerId": range(1, n+1),
    "Survived": np.random.randint(0, 2, n),
    "Pclass": np.random.randint(1,4,n),
    "Name": [f"{fake.last_name()}, {fake.first_name()}" for _ in range(n)],
    "Sex":[random.choice(["female", "male"]) for _ in range(n)],
    "Age": np.random.randint(5, 66, n),
    "SibSp": np.random.randint(0, 5, n),
    "Parch": np.random.randint(0, 6, n),
    "Ticket": np.random.randint(512580, 6000000, n),
    "Fare": np.random.randint(10, 151, n),
    "Cabin": [random.choice(["C12", "A54", "B98", "F5"]) for _ in range(n)],
    "Embarked": [random.choice(["S", "Q","C"]) for _ in range(n)],
})

df.to_csv("../data_retrain/new_data_in_progress.csv", index=False)

# PassengerId,Survived,Pclass,Name,Sex,Age,SibSp,Parch,Ticket,Fare,Cabin,Embarked
```
